```python
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/NATTAWUTI/Dataset/main/eec.csv'
df = pd.read_csv(url)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df['จังหวัด'].unique()
```

```python
##create group
grouped_df = df.groupby(by=["จังหวัด"])['ชื่อโรงงาน'].count().reset_index()
grouped_df.columns = ['Province' , 'Factory_Count']

#convert to dataframe
grouped_df1 = pd.DataFrame(grouped_df).sort_values(by='Factory_Count', ascending=True)

#create graph
fig = px.bar(grouped_df1, x='Factory_Count', y='Province', orientation='h', title='Total Factories by Province', text_auto=True,color="Factory_Count",color_continuous_scale='Viridis' ,)
fig.show()

```
